```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def show_image(image, title=None):
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    plt.title(title)
    plt.show()
```

Reference:
* convolution : https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.ndimage.filters.convolve.html
* median : https://docs.scipy.org/doc/scipy/reference/generated/scipy.ndimage.median_filter.html

### Gaussian noise

```python
# Read image
# img_noise = plt.imread('img_gaussian_noise.jpg')
from skimage import color
img_noise = color.rgb2gray(plt.imread('img_gaussian_noise2.png'))

print(img_noise.shape)
print(img_noise.dtype)
show_image(img_noise,'Gaussian noise')
show_image(img_noise[50:100,50:100])
```

```python
#apply gaussian filter
from skimage import filters
img_denoise_gaussian = filters.gaussian(img_noise, sigma=2)
show_image(img_denoise_gaussian, 'Gaussian filtered')
show_image(img_denoise_gaussian[50:100,50:100])
```

### Salt-and-pepper noise

```python
img_noise = plt.imread('img_salt_and_pepper.png')

print(img_noise.shape)
show_image(img_noise, 'Salt-and-pepper noise')
show_image(img_noise[50:100,50:100])
```

```python
#apply gaussian filter
from skimage import filters
img_denoise_gaussian = filters.gaussian(img_noise, sigma=2)
show_image(img_denoise_gaussian, 'Gassian filtered')
show_image(img_denoise_gaussian[50:100,50:100])
```

```python
#apply median filter
from scipy import ndimage as ndi
size = 5
img_denoise_median = ndi.median_filter(img_noise, size)

show_image(img_denoise_median, 'Median filtered')
show_image(img_denoise_median[50:100,50:100])
```
