```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_image(image, title=None):
    plt.imshow(image, cmap='gray', vmin=0, vmax=255)
    plt.axis('off')
    plt.title(title) 
```

### View R,G,B channels separately

```python
#view R,G,B channels separately 
titles = ['R', 'G', 'B']
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20,60))
for i in range(3):
    ax.ravel()[i].imshow(img_coffee[:,:,i], cmap='gray') 
    ax.ravel()[i].axis('off')
    ax.ravel()[i].set_title(titles[i] + ' channel')
plt.show()
```

### Count number of white tiles

```python
from skimage import data
img_checker = data.checkerboard()
```

```python
# 
img_checker.shape
img_checker.dtype
show_image(img_checker)
show_image(img_checker[20:30,20:30])
```

```python
# histogram
plt.hist(img_checker.ravel(), bins=256);
```

```python
#   255
img_bw = img_checker > 250  # threshold  1 
show_image(img_bw)
show_image(img_bw[20:30,20:30])
```

```python
#   
# image processing  connected component analysis
from skimage import measure
```

```python
img_label,num = measure.label(img_bw, return_num=True, connectivity=1)
```

```python
num
img_label[20:30,20:30]
```

### Convert to HSV
* hue : https://en.wikipedia.org/wiki/Hue

```python
from skimage import color
img_hsv = color.rgb2hsv(img_coffee)
```

```python
img_hsv.shape
```

```python
plt.hist(img_hsv[:,:,0].ravel()*360, bins=360)
```

```python
#shift the hue value [0-40] to [220-260]
img_h = (img_hsv[:,:,0]*360 + 220)%360 /360
img_hsv_new = np.stack((img_h,img_hsv[:,:,1],img_hsv[:,:,2]))
plt.hist(img_h.ravel()*360, bins=360)
```

```python
img_hsv_new = np.moveaxis(img_hsv_new,0,-1)
```

```python
show_image(color.hsv2rgb(img_hsv_new))
```

```python
show_image(color.hsv2rgb(img_hsv))
```
