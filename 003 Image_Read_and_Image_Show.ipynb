### Import modules จำเป็น

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### อ่าน grayscale image และแสดงโดยใช้ matplotlib

```python
#อ่านภาพเข้ามาในโปรแกรม

img_gray = plt.imread("img_grayscale.jpg")
print(type(img_gray))
print(img_gray.dtype)
print(img_gray.shape)
```

```python
#แสดงภาพ
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title('Gray Scale Image')
plt.show()
```

### อ่าน RGB image from data module และแสดงภาพ
* ภาพที่สามารถโหลดจาก module data สามารถเข้าไปดูได้ที่ https://scikit-image.org/docs/stable/api/skimage.data.html

```python
#โหลดภาพจรวด มาจาก module data ของ skimage
from skimage import data

img_rocket = data.rocket()

print(type(img_rocket))
print(img_rocket.dtype)
print(img_rocket.shape)
```

```python
#สร้าง function ไว้แสดงภาพ และลองแสดงภาพสี
def show_image(img, title=None):
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.title(title)
    plt.axis('off')

show_image(img_rocket, 'Rocket')
```

```python
#แสดง RGB channels แยกกัน
show_image(img_rocket[:,:,0], 'R-channel')
```

```python
show_image(img_rocket[:,:,1], 'G-channel')
```

```python
show_image(img_rocket[:,:,2], 'B-channel')
```

```python
#สามารถใช้ code นี้เพื่อแสดงทั้ง 3 channels บน figure เดียวกัน
titles = ['R', 'G', 'B']
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20,60))
for i in range(3):
    ax.ravel()[i].imshow(img_rocket[:,:,i], cmap='gray') 
    ax.ravel()[i].axis('off')
    ax.ravel()[i].set_title(titles[i] + '-channel')
plt.show()
```

### แบบฝึกหัด อ่านและแสดงรูปแก้วกาแฟจาก skimage.data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def show_image(img, title=None):
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.title(title)
    plt.axis('off')
    
#import module data จาก skimage
from skimage import data

#load รูปกาแฟ
img_coffee = data.coffee()

#แสดงรูปกาแฟ โดยระบุชื่อภาพเป็น coffee cup
show_image(img_coffee, "coffee cup")
```
