```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

###  grayscale image  matplotlib

```python
#

img_gray = plt.imread("img_grayscale.jpg")
print(type(img_gray))
print(img_gray.dtype)
print(img_gray.shape)
```

```python
#
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title('Gray Scale Image')
plt.show()
```

###  RGB image from data module 
*  module data  https://scikit-image.org/docs/stable/api/skimage.data.html

```python
#  module data  skimage
from skimage import data

img_rocket = data.rocket()

print(type(img_rocket))
print(img_rocket.dtype)
print(img_rocket.shape)
```

```python
# function  
def show_image(img, title=None):
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.title(title)
    plt.axis('off')

show_image(img_rocket, 'Rocket')
```

```python
# RGB channels 
show_image(img_rocket[:,:,0], 'R-channel')
```

```python
show_image(img_rocket[:,:,1], 'G-channel')
```

```python
show_image(img_rocket[:,:,2], 'B-channel')
```

```python
# code  3 channels  figure 
titles = ['R', 'G', 'B']
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20,60))
for i in range(3):
    ax.ravel()[i].imshow(img_rocket[:,:,i], cmap='gray') 
    ax.ravel()[i].axis('off')
    ax.ravel()[i].set_title(titles[i] + '-channel')
plt.show()
```
